#### 1) Delivery_time -> Predict delivery time using sorting time 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
data=pd.read_csv("C:\\Users\\SHARAN\\Desktop\\ExcelR\\Assignment\\Assignment_4 Simple linear regression\\delivery_time.csv")
data.rename(columns={"Delivery Time":"d","Sorting Time":"s"},inplace=True)
data.head(3)
```

```python
data.info()
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
sns.distplot(data.d)
```

```python
sns.distplot(data.s)
```

```python
boxplot=data[["d","s"]].boxplot()
```

```python
scatter=sns.scatterplot(x="s",y="d",data=data)
```

```python
data.corr()
```

```python
sns.heatmap(data.corr())
```

```python
model=smf.ols("d~np.log(s)",data=data).fit()
model.params
```

```python
model.summary()
```

```python
model.tvalues
```

```python
model.pvalues
```

```python
data1=data.iloc[:,1]
model.predict(data1)
```

```python
qqplot=sm.qqplot(model.resid,line='q')
plt.show()
```

#### 2) Salary_hike -> Build a prediction model for Salary_hike

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.regressionplots import influence_plot
import statsmodels.formula.api as smf
data=pd.read_csv("C:\\Users\\SHARAN\\Desktop\\ExcelR\\Assignment\\Assignment_4 Simple linear regression\\Salary_Data.csv")
data.head(3)
```

```python
data.info()
```

```python
data.rename(columns={"YearsExperience":"exp","Salary":"sal"},inplace=True)
data.head(3)
```

```python
data[['sal']].boxplot()
```

```python
data[['exp']].boxplot()
```

```python
sns.distplot(data.exp)
```

```python
sns.distplot(data.sal)
```

```python
sns.scatterplot(x='exp',y='sal',data=data)
```

```python
sns.regplot(x='exp',y='sal',data=data)
```

```python
data.corr()
```

```python
sns.heatmap(data.corr())
```

```python
model=smf.ols('sal~exp+np.square(exp)',data=data).fit()
model.params
```

```python
model.tvalues
```

```python
model.summary()
```

```python
data1=data.iloc[:,:1]
model.predict(data1)
```

```python
qqplot=sm.qqplot(model.resid,line='q')
plt.show()
```
